7.Simple calculator

```python
operation = input("Enter calculator symbol + - * / // %:")

def add(num):
    result = 0
    for i in num:
        result +=i
    return result

def sub(num):
    return num[0] - num[1]

def mul(num):
    result =1 
    for i in num:
        result*= i
    return result

def div(num):
    return num[0]//num[1]
def fdiv(num):
    return num[0]/num[1]
def mod(num):
    return num[0]%num[1]

if operation == '+' or operation == '-' or operation =='*' or operation =='/' or operation == '//' or operation=='%':
    operands = list(map(int,input("Enter the numbers separated by space:").split(" ")))
    if operation == '+':
        print("Addition:",add(operands))
    elif operation =='-':
        print("Substraction:",sub(operands))
    elif operation =='*':
        print("Multiplication:",mul(operands))
    elif operation =='//':
        print("Division:",div(operands))
    elif operation =='/':
        print("Floor Division: ",fdiv(operands))
    elif operation =='/':
        print("Modulus:",mod(operands))
else:
    print("Enter invalid operation...")
```

21.Armstrong number

```python
number = int(input())
copy = number
armstrong = 0
count = len(str(number))

while number >0:
    remainder = number %10
    armstrong += remainder**count
    number //=10

if armstrong == copy:
    print(f"{copy} is an Armstrong number")
else:
    print("not an armstrong number")
```

22.Armstong number in an interval

```python
start = int(input("Enter first number: "))
stop = int(input("Enter final number: "))

def isArmstrong(n):
    str_n = str(n)
    count = len(str_n)
    armstrong = 0
    for j in str_n:
        armstrong += int(j)**count
    if armstrong == n:
        return n
for i in range(start,stop+1):
    if isArmstrong(i):
        print(i,end=" ")
```

25.Find HCF or GCD

```python
nums = [int(x) for x in list(input().split(','))]

def gcd(nums):
    nums = sorted(nums)
    for j in range(nums[0],1,-1):
        boolean = False
        for i in nums:
            if i%j==0:
                boolean = True
            else:
                boolean = False
        if boolean == True:
            return j
print(gcd(nums))
```

39.Flatten list from nested list

```python
nested_list = [0,1,0.2,[3.1,4.2,[5,6,[70,80,[900,1000]]]]]
def flattened(num):
    flattened_list = []
    for i in num:
        if isinstance(i,list):
            flattened_list.extend(flattened(i))
        else:
            flattened_list.append(i)
        return flattened_list.append(i)
    print(flattened(nested_list))
```

43.Add 2 matrices

```python
m, n = map(int, input("Enter matrix structure in M*N: ").split())

print("Enter the elements of first matrix: ")
matrix1 = []
for i in range(m):
    row = []
    for j in range(n):
        k = int(input())
        row.append(k)
    matrix1.append(row)

print("Enter the elements of second matrix: ")
matrix2 = []
for i in range(m):
    row = []
    for j in range(n):
        k = int(input())
        row.append(k)
    matrix2.append(row)

# Matrix Addition
major = []
for i in range(m):
    minor = []
    for j in range(n):
        k = matrix1[i][j] + matrix2[i][j]
        minor.append(k)
    major.append(minor)

# Printing result
print("Resultant Matrix:")
for i in range(m):
    for j in range(n):
        print(major[i][j], end=" ")
    print()
```
